#Fungsi Parameter

```python
nama_siswa_1= "Daffanes"
nama_siswa_2= "Adzka"
nama_siswa_3= "Haidar"

print(nama_siswa_1, nama_siswa_2, sep="-")
```

```python
nama_siswa_1 = "Daffanes"
nama_siswa_2 = "Adzka"
nama_siswa_3 = "Haidar"

print(nama_siswa_3, nama_siswa_2, nama_siswa_1, nama_siswa_2, sep="*")
```

```python
print(nama_siswa_1)
print(nama_siswa_2)
```

```python
print(nama_siswa_1, nama_siswa_2)
```

```python
print(nama_siswa_1, end="-")
print(nama_siswa_2)
```
